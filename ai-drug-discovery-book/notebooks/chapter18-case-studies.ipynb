## Case Study 2: Kinase Inhibitors
- Example: Virtual screening and ML for kinase inhibitor discovery.

## Lessons Learned
- Data quality and integration are critical.
- Iterative cycles of prediction and validation improve outcomes.
- Collaboration between domain experts and data scientists is key.

- Demonstrates the use of network analysis [@kegg; @opentargets; @chembl] and GNNs [@hamilton2017representation; @zhang2021graph; @gnn_review] to prioritize compounds for a complex neurodegenerative disease.
- Integrates multi-agent workflow [@crewai; @multiagent_review] and biomedical data sources [@kegg; @chembl; @alphafold].
- See `demo_workflow.ipynb` for a step-by-step example.
