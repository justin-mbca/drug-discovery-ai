```python
# Evaluate validity of generated SMILES
from rdkit import Chem
smiles_list = ['CCO', 'invalid_smiles']
for s in smiles_list:
    mol = Chem.MolFromSmiles(s)
    print(f'{s}: {'valid' if mol else 'invalid'}')
```
